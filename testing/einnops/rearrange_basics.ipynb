```python
# Create a sample 2D array
x = np.arange(12).reshape(3, 4)
print("Original array:\n", x)
```

```python
# Rearrange: flatten the array
flat = rearrange(x, 'h w -> (h w)')
print("\nFlattened array:\n", flat)
```

```python
# Rearrange: swap axes
swapped = rearrange(x, 'h w -> w h')
print("\nSwapped axes:\n", swapped)
```

```python
# Rearrange: split each row into two columns
split = rearrange(x, 'h (w1 w2) -> h w1 w2', w1=2)
print("\nSplit columns:\n", split)
```
